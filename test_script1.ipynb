```python
from Bio import Entrez
import datetime
import csv
import sys
Entrez.email = "kevin.esoh@students.jkuat.ac.ke"
api_key = input("Please enter API key (hit enter if you don't have it): ").lower()
currentDate = datetime.date.today()
currentTime = datetime.datetime.now()

def main():
    dataBase = getDataBase()
    searchTerm = getSearchTerm()
    dbSearch(dataBase, searchTerm)
    with open("idList.txt", "r+") as newFile:
        newIdList = csv.reader(newFile)
        for eachID in newIdList:
            idList = ','.join(eachID)
    dbSummary(dataBase, idList)
    interestingIDs = input("Please enter ID(s) for which you want abstract(s) separated by comma (e.g. 30986542,30975198,30894432): ")
    getAbstract(dataBase, interestingIDs)
    return

def getDataBase():
    dataBase = input("Please provide the NCBI database you wish to search: ").lower()
    return dataBase

def getSearchTerm():
    searchTerm = input("Enter you search word(s): ").lower()
    return searchTerm

def dbSearch(dataBase, searchTerm):
    handle = Entrez.esearch(db=dataBase, retmax=10, term=searchTerm, rettype="uilist", retmode="xml")
    record = Entrez.read(handle)
    #record2 = Entrez.read(handle, validate=True)
    handle.close()
    
    print("Your query: " + record["QueryTranslation"])
    print(" ")
    print("Total count: " + record["Count"] + \
          " Accessed: " + currentDate.strftime('%a %d %B, %Y') + \
          " At: " + currentTime.strftime('%H:%M:%S'))
    idList = record["IdList"]
    print(" ")
    print(idList)
    newList = open("idList.txt", "w+")
    newList.write(',' .join(idList))
    newList.close()
    #print(idList)
    #print(',' .join(idList))
    #print(" ")
    #print(record["Title"])
    return

def dbSummary(dataBase, idList=[]):
    handle = Entrez.esummary(db=dataBase, id=idList, retmode="xml")
    #records = Entrez.read(handle)
    records = Entrez.parse(handle)
    for record in records:
        print(" ")
        #print(record)
        print(record['Id'] + " | " + \
                record['Title'] + " | " + \
                record['AuthorList'][0] + " | " + \
                record['PubDate'])
    handle.close()
    print(" ")
    return

def getAbstract(dataBase, interestingIDs=[]):
    handle = Entrez.efetch(db=dataBase, id=interestingIDs, retmode="xml")
    records = Entrez.read(handle)
    print(" ")
    print(records)
    print(" ")
    #print(records.items())
    print(" ")
    #print(records.keys())
    print(" ")
    print(records['PubmedArticle'])
    print(" ")
    #print(records.values())
    newDict = dict(records['PubmedArticle'])
    print(" ")
    print(newDict.keys())
    print(newDict.values())
    print(newDict.items())
    print(" ")
    print(newDict['PubmedData'])
    try:
        for record in records:
            print(" ")
            print(record)
            print('' .join(record))
    except:
        error = sys.exc_info()[0]
        print("Sorry an error occurred!")
        print(error)
    handle.close()
    print(" ")
    return
main()
```
