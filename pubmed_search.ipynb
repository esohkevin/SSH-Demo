```python
from Bio import Entrez
import datetime
import csv
import sys
Entrez.email = "kevin.esoh@students.jkuat.ac.ke"
apiKey=''.join(open("api_key.txt", "r"))
api_key = apiKey.lower()
#api_key = input("Please enter API key (hit enter if you don't have it): ").lower()
currentDate = datetime.date.today()
currentTime = datetime.datetime.now()

def main():
    dataBase = getDataBase()
    searchTerm = getSearchTerm()
    dbSearch(dataBase, searchTerm)
    with open("idList.txt", "r+") as newFile:
        newIdList = csv.reader(newFile)
        for eachID in newIdList:
            idList = ','.join(eachID)
    dbSummary(dataBase, idList)
    getAbstract(dataBase, idList)
    return

def getDataBase():
    dataBase = input("Please provide the NCBI database you wish to search: ").lower()
    return dataBase

def getSearchTerm():
    searchTerm = input("Enter you search word(s): ").lower()
    return searchTerm

def dbSearch(dataBase, searchTerm):
    handle = Entrez.esearch(db=dataBase, retmax=5, term=searchTerm, retmode="uid")
    record = Entrez.read(handle)
    totCount = int(record["Count"])
    handle.close()
    handle = Entrez.esearch(db=dataBase, retmax=totCount, term=searchTerm, retmode="uid")
    record = Entrez.read(handle)
    #record2 = Entrez.read(handle, validate=True)
    handle.close()
    
    print("Your query: " + record["QueryTranslation"])
    print(" ")
    print("Total count: " + record["Count"] + \
          " Accessed: " + currentDate.strftime('%a %d %B, %Y') + \
          " At: " + currentTime.strftime('%H:%M:%S'))
    idList = record["IdList"]
    print(" ")
    print(','.join(idList))
    newList = open("idList.txt", "w+")
    newList.write(',' .join(idList))
    newList.close()
    saveFile = "abstracts.txt"
    newFile = open(saveFile, "w+")
    newFile.write("#Database: " + dataBase + "\n" + "#Query: " + record["QueryTranslation"] + "\n" + \
                  "#Accessed: " + currentDate.strftime('%a %d %B, %Y') + " | " + \
                  currentTime.strftime('%H:%M:%S') + "\n" + "#Count: " + record['Count'] + "\n")
    newFile.close()
    #print(idList)
    #print(',' .join(idList))
    #print(" ")
    #print(record["Title"])
    return

def dbSummary(dataBase, idList=[]):
    handle = Entrez.esummary(db=dataBase, id=idList, rettype="xml")
    #records = Entrez.read(handle)
    records = Entrez.parse(handle)
    # handle errors
    try:
        for record in records:
            print(" ")
            #print(record)
            print(record['Id'] + " | " + \
                    record['Title'] + " | " + \
                    record['AuthorList'][0] + " | " + \
                    record['PubDate'] + " | " + \
                    record['DOI'])
    except:
        error = sys.exc_info()[0]
        print("Sorry an error occurred!")
        print(error)
        #sys.exit() #exit program if error occurs
    handle.close()
    print(" ")
    return

def getAbstract(dataBase, idList=[]):
    handle = Entrez.efetch(db=dataBase, id=idList, rettype="abstract", retmode="text")
    with open("abstracts.txt", "a") as newFile:
        newFile.write(''.join(handle))
        newFile.close()
    try:
        with open("abstracts.txt", "r") as medlineFile:
            print(csv.reader(''.join(medlineFile)))
    except:
        error = sys.exc_info()[0]
        print("Something went wrong!")
        print(error)
    try:
        for record in records:
            print(" ")
            print(record)
            print('' .join(record))
    except:
        error = sys.exc_info()[0]
        print("Sorry an error occurred!")
        print(error)
    handle.close()
    print(" ")
    return
main()
```
